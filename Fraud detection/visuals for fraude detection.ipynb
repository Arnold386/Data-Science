```python
import pandas as pd
from sklearn import preprocessing
```

```python
df = pd.read_csv('fraude_detection.csv', parse_dates=['approveddate','creationdate'])
```

```python
df.head(20)
```

```python
df1 = df.drop(['customerid','referredby','systemloanid'], axis='columns')
```

```python
df1.head()
```

```python
df1['approved_year'] = df1['approveddate'].dt.year
df1['approved_month'] = df1['approveddate'].dt.month
df1['approved_day'] = df1['approveddate'].dt.day
df1['approved_day_of_week'] = df1['approveddate'].dt.dayofweek
df1['approved_hour'] = df1['approveddate'].dt.hour
```

```python
df1.head()
```

```python
df2 = df1.drop(['approveddate'], axis='columns')
df2.head()
```

```python
df2['created_year'] = df1['creationdate'].dt.year
df2['created_month'] = df1['creationdate'].dt.month
df2['created_day'] = df1['creationdate'].dt.day
df2['created_day_of_week'] = df1['creationdate'].dt.dayofweek
df2['created_hour'] = df1['creationdate'].dt.hour
```

```python
df3 = df2.drop('creationdate', axis='columns')
df3.head()
```

```python
le = preprocessing.LabelEncoder()
q = le.fit_transform(df3['good_bad_flag'])
```

```python
df4 = pd.DataFrame(q, columns=['Fraud = 0, no Fraud = 1'])
```

```python
df4.head()
```

```python
finaldf = pd.concat([df3,df4], axis='columns')
```

```python
finaldf.head()
```

```python
f = finaldf.drop('good_bad_flag', axis='columns')
```

```python
f.head()
```

```python
f.head()
```

```python
g = f.set_index('loannumber')
```
