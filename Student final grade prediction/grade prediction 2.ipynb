```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle
```

```python
df = pd.read_csv('student-mat.csv', sep = ";")
df.head(10)
```

```python
df = df[["G1", "G2", "G3", "studytime", "failures", "absences"]]
#to be predicted final grade G3
predict = "G3"
#we now create arrays.
x = np.array(df.drop([predict], 1))
y = np.array(df[predict])

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size = 0.1)
```

```python
linear = linear_model.LinearRegression()
```

```python
linear.fit(x_train, y_train)
acc = linear.score(x_test, y_test)# acc is accuracy
```

```python
print(acc)
```

```python
#to view constants used to generate linear line
print('coefficient: \n', linear.coef_)#each slope value
print('intercept; \n', linear.intercept_)#intercept of the line
```

```python
# predicting on specific students
predictions = linear.predict(x_test)#gets a list of all predictions
print('Predicted result......'+'........Actual result')    
for x in range(len(predictions)):
    print(predictions[x], x_test[x], y_test[x])#printin the predicted value for test samples alongside the actual score they got
```
